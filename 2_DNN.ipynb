## Building the Deep Learning model

```python
# Charging the necessary dependencies
import numpy as np                              # Library for using arrays and numeric calculus
import pandas as pd
import matplotlib.pyplot as plt                 # Library for plotting
import seaborn as sns                           # Library for providing style to the charts
from sklearn import preprocessing               # Library for preprocessing data
from sklearn.metrics import r2_score
from keras.models import Sequential
from keras.layers import Dense
%matplotlib inline
```

```python
# Charging the data
df = pd.read_csv("Dataset_predicción.csv")
df
```

```python
# Cretaing the model architecture
input_size = 1
Output_size = 1
hidden_layer_size = 100
model = Sequential([
        Dense(hidden_layer_size, activation = 'relu', kernel_initializer = 'he_uniform', input_dim= input_size),
        Dense(hidden_layer_size, activation = 'relu', kernel_initializer = 'he_uniform'),
        Dense(hidden_layer_size, activation = 'relu', kernel_initializer = 'he_uniform'),
        Dense(Output_size)
])
# Dense = capas ocultas 
```

# Sección nueva

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
# Compiling the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
# mse = mean squared error
# mae = mean absolute error

# se pueden usar varios al menos en metrics
```

```python
model.summary()
```

```python
df.index
```

Codificación ordinal: Si la variable de fecha tiene un orden natural (por ejemplo, días consecutivos), puedes codificarla de manera ordinal asignando números enteros a cada valor de fecha en función de su orden. Esto permite que el modelo aprenda la relación de orden entre las fechas.

```python
# Training the model
modelo = model.fit(df.index, df['Adj Close'], epochs=500, batch_size=10)
# epochs = iteraciones
```

```python
loss = modelo.history['loss']
epochs = range (len(loss))
```

```python
plt.plot(epochs, loss, 'r--', label = 'Training loss')
plt.title('Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid()
plt.savefig("loss.pdf")
```

```python
# Crear conjunto de datos de prueba
test_df = pd.DataFrame({'Date': ['2022-01-15', '2022-01-16'], 'Adj Close': [200.07, 300]})
```

```python
# Preprocesamiento
test_df = test_df.set_index('Date')
```

```python
# Realizar predicciones
predictions = model.predict(test_df)

# Imprimir predicciones
print(predictions)
```
